```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "/content/drive/MyDrive/TUGAS SLO/data0-SZL.csv"
data = pd.read_csv(path, header=None)
```

## **Nilai Rata-rata, Standar deviasi, Skewness, Median, Energi, Min, Max**

```python
mean = data.mean()
std = data.std()
skewness = data.skew()
median = data.median()
energy = (data ** 2).sum()
min_val = data.min()
max_val = data.max()

statistics_df = pd.DataFrame({
    'Mean': mean,
    'Standard Deviation': std,
    'Skewness': skewness,
    'Median': median,
    'Energy': energy,
    'Min': min_val,
    'Max': max_val
})

# Display the statistics DataFrame
statistics_df
```

## **Random Sample Multiplot 2x3 dari Data Sinyal**

```python
selected_columns = data.sample(6, axis=1)

# Plot sinyal dengan multiplot 2x3
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

for i, ax in enumerate(axes.flatten()):
    col = selected_columns.columns[i]
    ax.plot(selected_columns[col])
    ax.set_title(col)

plt.tight_layout()
plt.show()
```

## **Scatter Plot Mean & Standar Deviasi**

```python
plt.figure(figsize=(10, 6))

for i in range(0, len(mean), 10):
    plt.scatter(mean[i:i+10], std[i:i+10], label=f'Data Points {i+1}-{i+10}')

plt.xlabel('Mean')
plt.ylabel('Standar Deviasi')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Scatter Plot')

plt.show()
```

## **Scatter Plot Skewness & Median**

```python
plt.figure(figsize=(10, 6))

for i in range(0, len(skewness), 10):
    plt.scatter(skewness[i:i+10], median[i:i+10], label=f'Data Points {i+1}-{i+10}')

plt.xlabel('Skewness')
plt.ylabel('Median')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Scatter Plot')

plt.show()
```
